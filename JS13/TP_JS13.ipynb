# Tugas Praktikum
Gunakan JST untuk klasifikasi angka tulisan tangan (MNIST).

Langkah:

Load dataset MNIST dari Keras.

Bangun model dengan 2 hidden layer.

Latih model dan evaluasi akurasi.

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# 1. Load dataset MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisasi data (0-255 → 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 2. Bangun model JST
model = Sequential([
    # Ubah gambar 28x28 menjadi vektor
    Flatten(input_shape=(28, 28)),
    # Hidden layer 1
    Dense(128, activation='relu'),
    # Hidden layer 2
    Dense(64, activation='relu'),
    # Output layer (10 kelas)
    Dense(10, activation='softmax')
])

# 3. Kompilasi model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 4. Latih model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.1)
# model.fit(X_train, y_train, epochs=50, batch_size=8)

# 5. Evaluasi model
loss, acc = model.evaluate(X_test, y_test)
print(f"Akurasi pada data uji: {acc:.4f}")
```

Coba dengan beberapa parameter lain:
- Ubah jumlah neuron di hidden layer (misal: 256 dan 128).
- Tambahkan satu hidden layer lagi.
- Bandingkan akurasi dan waktu pelatihan.
- Eksperimen dengan fungsi aktivasi Sigmoid vs ReLU.

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# 1. Load dataset MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisasi data (0-255 → 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 2. Bangun model JST relu dengan 3 hidden layer
model_relu = Sequential([
    # Ubah gambar 28x28 menjadi vektor
    Flatten(input_shape=(28, 28)),
    # Hidden layer 1
    Dense(256, activation='relu'),
    # Hidden layer 2
    Dense(128, activation='relu'),
    #hidden layer tamnahan
    Dense(64, activation='relu'),
    # Output layer (10 kelas)
    Dense(10, activation='softmax')
])

# 2. Bangun model JST sigmoid dengan 3 hidden layer
model_sigmoid = Sequential([
    # Ubah gambar 28x28 menjadi vektor
    Flatten(input_shape=(28, 28)),
    # Hidden layer 1
    Dense(256, activation='sigmoid'),
    # Hidden layer 2
    Dense(128, activation='sigmoid'),
    #hidden layer tamnahan
    Dense(64, activation='sigmoid'),
    # Output layer (10 kelas)
    Dense(10, activation='softmax')
])

# 3. Kompilasi model
model_relu.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model_sigmoid.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 4. Latih model relu
mulai_relu = tf.timestamp()
model_relu.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.1)
selesai_relu = tf.timestamp()

# 4. Latih model sigmoid
mulai_sigmoid = tf.timestamp()
model_sigmoid.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.1)
selesai_sigmoid = tf.timestamp()

# 5. Evaluasi model
loss_relu, acc_relu = model_relu.evaluate(X_test, y_test)
loss_sigmoid, acc_sigmoid = model_sigmoid.evaluate(X_test, y_test)

#PERBANDINGANNNNN
print("\n=== PERBANDINGAN ===")
print(f"Sigmoid - Loss: {loss_sigmoid:.4f}, Akurasi: {acc_sigmoid:.4f}, Waktu Latih: {selesai_sigmoid - mulai_sigmoid:.2f} detik")
print(f"ReLU    - Loss: {loss_relu:.4f}, Akurasi: {acc_relu:.4f}, Waktu Latih: {selesai_relu - mulai_relu:.2f} detik")
print(f"\nSelisih Akurasi: {abs(acc_relu - acc_sigmoid):.4f}")
```
